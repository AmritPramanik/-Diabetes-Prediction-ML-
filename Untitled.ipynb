```python
X = Diabetes_df.iloc[:,0:-1]
X.head()
```

```python
y = Diabetes_df.iloc[:,-1]
y
```

# Data Standarization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(X)
X
```

# Split Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
X_train.shape,X_test.shape
```

# Model Train & Test

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear')
classifier.fit(X_train,y_train)
```

```python
X_pred = classifier.predict(X_train)
```

```python
y_pred = classifier.predict(X_test)
```

# Model Evaluation

<b><h3>Accuracy Score

```python
from sklearn.metrics import accuracy_score
```

```python
train_accuracy_score = accuracy_score(y_train,X_pred)
print("Aaccuracy score on Test data: ",train_accuracy_score)
```

```python
test_accuracy_score = accuracy_score(y_test,y_pred)
print("Aaccuracy score on Test data: ",test_accuracy_score)
```

Prediction Aaccuracy score is around 78% prity much good <br>

# Making a Predctive System

```python
def input_information():
    Pregnancies = int(input("Pregnancies : "))
    Glucose = int(input("Glucose : "))
    BloodPressure = int(input("BloodPressure : "))
    SkinThickness = int(input("SkinThickness : "))
    Insulin = int(input("Insulin : "))
    BMI = float(input("BMI : "))
    DPF = float(input("DiabetesPedigreeFunction : "))
    Age = int(input("Age : "))

    data =  (Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DPF,Age)
    return data
```

```python
def check_prediction():
    input_data = input_information()
    input_data = np.array(input_data)
    input_data_reshape = input_data.reshape(1,-1)

    # Standarize the data
    std_input = scaler.transform(input_data_reshape)
    prediction = classifier.predict(std_input)

    return prediction
```

```python
if(check_prediction()):
    print("Diabetic case")
    
else : 
    print("Non Diabetic case")
```
